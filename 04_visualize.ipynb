```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Heat map


```python
flights = sns.load_dataset("flights")
flights
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
tipscorr = tips[["total_bill", "tip", "size"]]
tipscorr
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr(), annot=True)
```

### Cluster map


```python
sns.clustermap(tipscorr.corr())
```

### Pivot table Heat map


```python
pvtflights = flights.pivot(index="year", columns="month", values="passengers")
pvtflights
```

```python
sns.heatmap(pvtflights)
```
